### 1.5 전체를 상대로!
: <u>scaling 수치가 NaN인 경우 존재할거야  - "keyword가 단 하나의 카테고리에서만 나타난 경우"

: 오히려 잘됐어. NaN인건 제외하고 수행하자 - keyword내에서 카테고리 할 대상들이 없는거니깐?

```python
# 전체를 상대로!

keyword_scaling = keyword_category.groupby('keyword')\
                  .apply(lambda df : (df.cnt-df.cnt.min())/(df.cnt.max()-df.cnt.min()))\
                  .reset_index().set_index("level_1")

```

### 1.6 <u>keyword마다 '빈도' 도 넣어줄게 ('freq')<u>

<상대빈도와 scaling 수치가 어떻게 해석될까?> - 어느정도 비슷한 의미를 가질거야! <br/>
  예를 들어, 어느 키워드 내에서 min-max scaling 수치가 1이더라도 freq는 다양하게 나타날 수 있어.(키워드 내에서 카테고리 분포가 고르면 freq는 별로 차이가 안날거고(그때에는 scaling 수치도 다들 크게 차이나지 않겠지) - 그런 의미에서 비슷하다!)   

```python
keyword_freq = keyword_category_.groupby('keyword').apply(lambda x: x.cnt/x.cnt.sum()).reset_index()
```

```python
keyword_freq.set_index('level_1', inplace=True)
```

```python
keyword_freq.columns=['keyword', 'freq']
```
